```python
import json
import numpy as np
```

```python
with open("../data_relaxed/relaxed_filtered/passed_segments.json", "r") as f:
    data = json.load(f)
```

```python
total_time = 0

for item in data:
    total_time += (item["end"] - item["start"])
```

```python
print(f"~{round(total_time / 60 / 60, 1)} hours of training audio")
```

```python
print(f"Avg segment duration: {round(total_time / len(data), 2)}")
```
